```python
newResults = Results.filter(['Property_id','Stay_date','Booking_status','Room_name','Room_details','Room_rate','Channel'], axis=1)
newResults.head()
```

```python
newResults = newResults[(newResults.Booking_status == 'COMPLETED') | (newResults.Booking_status == 'CONFIRMED')]
newResults.head()
```

```python
newResults = newResults[(newResults.Stay_date >= '2015-01-01') & (newResults.Stay_date <= '2015-12-31')]
newResults.head()
```

```python
newResults['Stay_date'] = pd.to_datetime(newResults.Stay_date)
```

```python
newResults.head()
```

```python
res = pd.merge(left=newResults, left_on='Stay_date',right=scrapeResult, right_on='Date')
```

```python
res.head()
```

```python
res1 = res.drop_duplicates()
```

```python
res1.head()
```

```python
res1 = res1.reset_index()
```

```python
res1.head()
```

```python
del res1['index']
```

```python
res1.head()
```

```python
del res1['Date']
```

```python

res1.head()
```

```python
res1['room_count'] = res1.groupby(['Property_id','Stay_date'])['Room_rate'].transform('count')

res2 = res1.drop_duplicates()
res2
```

```python

res2['totalRevenue'] = res2.groupby(['Property_id','Stay_date'])['Room_rate'].transform('sum')
res3 = res2.drop_duplicates()
res3
```

```python
# test = res3[(res3.Property_id == "BB") & (res3.Stay_date=="2015-01-01")]
# test
```

```python
res3['ADR'] = res3.apply(lambda res3: (res3['totalRevenue'] / res3['room_count'] ) , axis=1)
```

```python
res4 =res3
res4
```

```python

res4['directChannelCount'] = res4.groupby(['Channel','Stay_date'])['Channel'].transform(lambda res4: res4[res4.str.contains('DIRECT')].count())
```

```python
res4
```

```python
res4['travelAgentsChannelCount'] = res4.groupby(['Channel','Stay_date'])['Channel'].transform(lambda res4: res4[res4.str.contains('TRAVEL AGENTS')].count())
```

```python
res4
```

```python
res4['corporatesChannelCount'] = res4.groupby(['Channel','Stay_date'])['Channel'].transform(lambda res4: res4[res4.str.contains('CORPORATES')].count())
```

```python
res4
```

```python
res4['OTAChannelCount'] = res4.groupby(['Channel','Stay_date'])['Channel'].transform(lambda res4: res4[res4.str.contains('OTA')].count())
```

```python
# Our numerical workhorses
import numpy as np
import pandas as pd

# Import Bokeh modules for interactive plotting
import bokeh.charts
import bokeh.charts.utils
import bokeh.io
import bokeh.models
import bokeh.palettes
import bokeh.plotting

# Display graphics in this notebook
bokeh.io.output_notebook()
```

```python
import bokeh.plotting as plt
from bokeh.plotting import figure, output_file, show, ColumnDataSource
from bokeh.models import HoverTool
from bokeh.models.formatters import DatetimeTickFormatter
```

```python
property_id_BB = pd.read_excel('C:/Users/Amulya/Desktop/EventsDataAnalysis/Task2/Results/GroupByPropertyID/BB.xlsx')
property_id_BB = property_id_BB.reset_index()
property_id_BB = property_id_BB.sort_values(by = 'Stay_date', ascending=True)
property_id_BB1 = property_id_BB.values.T.tolist()
```

```python
source = ColumnDataSource(
        data=dict(
            x=property_id_BB1[2],
            y=property_id_BB1[6],
            desc=property_id_BB1[4],
            date=[x.strftime("%Y-%m-%d") for x in property_id_BB['Stay_date']]
        )
    )

hover = HoverTool(
        tooltips=[
            ("Date","@date"),
            ("Event", "@desc"),
            ("Total Revenue","@y{1.11}")
        ]
    )

p = plt.figure(x_axis_type="datetime",width=800, height=500, tools=[hover],
           title="Graph of StayDate vs TotalRevenue for BB (hover to view events)",x_axis_label = "Stay Date",
               y_axis_label = "Total Revenue")
p.left[0].formatter.use_scientific = False
p.line('x', 'y', line_width=2, source=source)   
show(p)
```

```python
source = ColumnDataSource(
        data=dict(
            x=property_id_BB1[2],
            y=property_id_BB1[7],
            desc=property_id_BB1[4],
            date=[x.strftime("%Y-%m-%d") for x in property_id_BB['Stay_date']]
        )
    )

hover = HoverTool(
        tooltips=[
            ("Date","@date"),
            ("Event", "@desc"),
            ("ADR","@y{1.11}")
        ]
    )

p = plt.figure(x_axis_type="datetime",width=800, height=500, tools=[hover],
           title="Graph of StayDate vs ADR for BB (hover to view events)",x_axis_label = "Stay Date",
               y_axis_label = "ADR")
p.left[0].formatter.use_scientific = False
p.line('x', 'y', line_width=2, source=source)   
show(p)
```
