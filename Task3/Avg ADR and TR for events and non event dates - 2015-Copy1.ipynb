```python
newResults = Results.filter(['Property_id','Stay_date','Booking_status','Room_name','Room_details','Room_rate','Channel'], axis=1)
newResults.head()
```

```python
newResults = newResults[(newResults.Booking_status == 'COMPLETED') | (newResults.Booking_status == 'CONFIRMED')]
newResults.head()
```

```python
res = pd.merge(left=newResults, left_on='Stay_date',right=scrapeResult, right_on='Stay_date', how='left')

#res = pd.merge(newResults, scrapeResult, on='Stay_date', how='left')
```

```python
res = res.sort_values(by = 'Stay_date', ascending=True)
res.head()
```

```python
res1 = res.drop_duplicates()
```

```python
res1.head()
```

```python
res1 = res1.reset_index()
```

```python
res1.head()
```

```python
del res1['index']
```

```python
res1.head()
```

```python
res1 = res1.fillna('-')
```

```python
events = res1[(res1.Event != '-') ]
events
```

```python
events['totalRevenue'] = events.groupby(['Property_id','Stay_date'])['Room_rate'].transform('mean')
# events
```

```python
# events_tr = events["totalRevenue"].mean()
# events_tr
```

```python
events['room_count'] = events.groupby(['Property_id','Stay_date'])['Room_rate'].transform('count')
events.head()
```

```python
events['ADR'] = events.apply(lambda events: (events['totalRevenue'] / events['room_count'] ) , axis=1)
```

```python
# Our numerical workhorses
import numpy as np
import pandas as pd

# Import Bokeh modules for interactive plotting
import bokeh.charts
import bokeh.charts.utils
import bokeh.io
import bokeh.models
import bokeh.palettes
import bokeh.plotting

# Display graphics in this notebook
bokeh.io.output_notebook()
```

```python
import bokeh.plotting as plt
from bokeh.plotting import figure, output_file, show, ColumnDataSource
from bokeh.models import HoverTool
from bokeh.models.formatters import DatetimeTickFormatter
```

```python
property_id_BB = events[(events.Property_id == 'BB')]
property_id_BB = property_id_BB.reset_index()
property_id_BB = property_id_BB.sort_values(by = 'Stay_date', ascending=True)
property_id_BB1 = property_id_BB.values.T.tolist()
```

```python
property_id_BB16 = pd.read_excel('Results/BB16.xlsx')
property_id_BB16 = property_id_BB16.reset_index()
property_id_BB16 = property_id_BB16.sort_values(by = 'Stay_date', ascending=True)
property_id_BB116 = property_id_BB16.values.T.tolist()
```

```python
source = ColumnDataSource(
        data=dict(
            x=property_id_BB1[2],
            y=property_id_BB1[8],
            desc=property_id_BB1[7],
            date=[x.strftime("%Y-%m-%d") for x in property_id_BB['Stay_date']]
        )
    )

sourceA = ColumnDataSource(
        data=dict(
            x=property_id_BB116[2],
            y=property_id_BB116[8],
            desc=property_id_BB116[7],
            date=[x.strftime("%Y-%m-%d") for x in property_id_BB16['Stay_date']]
        )
    )

hover = HoverTool(
        tooltips=[
            ("Date","@date"),
            ("Event", "@desc"),
            ("Total Revenue","@y{1.11}")
        ]
    )

p = plt.figure(x_axis_type="datetime",width=800, height=500, tools=[hover],
           title="Graph of StayDate vs TotalRevenue for BB (hover to view events)",x_axis_label = "Stay Date",
               y_axis_label = "Total Revenue")
p.left[0].formatter.use_scientific = False
p.line('x', 'y', line_width=2, source=source, legend="2015", color="green")   
p.line('x', 'y', line_width=2, source=sourceA, legend="2016", color="blue")   
show(p)
```
